```python
# 5.
n = 3
p = 4
m = 5
array1 = np.random.rand(n, p)
array2 = np.random.rand(p, m)
print("Array 1 (n x p):")
print(array1)
print("Array  (p x m):")
print(array2)
print("Dot product of Array 1 and Array 2:")
dot_product = np.dot(array1, array2)

print(dot_product)
```
